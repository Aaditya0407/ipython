### 하드웨어 수준

- 현재 하드웨어 보안 레벨은 2000년대 초중반 PC 시장의 보안 레벨과 비슷하다고 한다. 그냥 간단한 것으로도 뻥뻥 뚫리기 때문에
- 이미 PC 시장에서 공격했던 취약점들이 모두 모바일로 옮겨갔고 IoT로도 넘어올거다. 그 얼마간의 기간동안은 그야말로 무주공산. 그냥 깃발 꼽고 내 땅이라고 하면 된다.

### 트레이닝을 통해 배우게 되는 것들

- 1일차: 소개, 전자/전기 기초, AVR 프로그래밍
- 2일차: 하드웨어 프로토콜, UART 해킹, 전기신호 Sniffer 개발 실습
- 3일차: ARM Exploitation, IPTIME 유무선 공유기 해킹 실습
- 4일차: Flash Memory Dump 실습
- 5일차: JTAG Port를 이용한 Hacking, Firmware Dump, AVR 바이너리 리버싱

- 찾는 곳이 많다?!

## 2일차

- 하드웨어 통신 프로토콜에 대해서 설명
- 동기와 비동기의 차이. 크.. 학부 시절 생각나더라. 교수님 얼굴도 생각이 났다.
- UART라고 하니까 뭔가 있어 보이는데 RS232다. 별거 없음.
- Logic Analyzer로 직접 키보드 연결해서 A 신호 찾기
  - A는 아스키코드로 0x41, 65다. 그런데 키보드 레벨에서는 다른 신호들로 정의가 되어 있다. 즉, 레이어 레벨이 다르다. 이 부분 주의할 것
  - 이걸 보니까 저번에 출입문 통제 시스템 하던게 떠올랐음. 그냥 열리는 패킷 복사해서 다시 보내면 게임 오버.. 엄청 취약.
- UART를 쓰게 된 이유를 아주 잘 설명해주심. 개발자들의 개발 고충 때문에 생겨난 비화?!
- 장비를 뜯어보고 4개의 핀이 달려있는 곳을 찾아라. 왠만하면 UART 포트다.

### 펌웨어를 획득하는 6가지 방법

1. 제조사에서 공개하는 펌웨어 다운로드
2. 자동/수동 업데이트가 될 때 패킷 스니핑
3. UART 포트 접속
4. 논리적 취약점을 이용하여 Shell 접근 권한 획득 후 추출(partition dump, /dev/mtdblock)
5. Flash Memory 덤프
6. JTAG 포트 접속

## 5일차

- JTAG: mironette(꼭두각시). CPU를 마음대로 조정할 수 있다.
- 한 마디로 4개 선으로 모든 동작을 제어할 수 있다. 라는게 핵심
  - TDI: Test Data In
  - TDO: Test Data Out
  - TMS: Test Mode Select
  - TCK: Test Clock
- 여기에서 Rising edge와 Falling edge 부분 나옴. 클럭과의 상관관계도 잘 이해해야 됨
- 초기화, DR영역, IR영역을 모두 함수로 짜신 옆에 분이 있었는데 잘 짜시더라;; 한 눈에 쏘옥 알아보기 쉽게.
- 역시 하드웨어라 비트단위 연산을 많이 해야 함
- 여기서 또 ID 레지스터를 읽는데 데이터시트를 보고 찾아야 한다. 그만큼 데이터시트를 보는 방법도 중요함
