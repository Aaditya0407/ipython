## LED 자동으로 켜지기 && Switch로 LED 켜고 끄기

```c


void setup() {
  pinMode(12, OUTPUT);
  pinMode(13, OUTPUT);
  pinMode(2, INPUT_PULLUP);
  pinMode(3, INPUT_PULLUP);
  
  attachInterrupt(0, led_on, FALLING); // pin2
  attachInterrupt(1, led_off, FALLING); // pin3
}

void loop() {
  digitalWrite(13, OUTPUT);
  delay(1000);
  digitalWrite(13, LOW);
  delay(1000);
}

void led_on()
{
  digitalWrite(12, HIGH); 
}

void led_off()
{
  digitalWrite(12, LOW);
}
```

## Timer Interrupt 이용한 LED 2개 켜고 끄기

```c
#include <MsTimer2.h>

void led_onoff()
{
  if(digitalRead(12) == HIGH) digitalWrite(12, LOW);
  else digitalWrite(12, HIGH);
}

void setup()
{
  pinMode(12, OUTPUT);
  pinMode(13, OUTPUT);

  MsTimer2::set(400, led_onoff); // 500ms period
  MsTimer2::start();
}

void loop()
{
  digitalWrite(13, HIGH);
  delay(300);
  digitalWrite(13, LOW);
  delay(300);
}
```

## Timer 100으로 세서

```c
#include <MsTimer2.h>

int count = 0;

void led_onoff()
{
  count++;
}

void setup()
{
  pinMode(12, OUTPUT);
  pinMode(13, OUTPUT);

  MsTimer2::set(100, led_onoff); // 500ms period
  MsTimer2::start();

}

void loop()
{
  if(count%3)
  {
    if(digitalRead(12) == HIGH) digitalWrite(12, LOW);
    else digitalWrite(12, HIGH); 
  }
  if(count%4)
  {
    if(digitalRead(13) == HIGH) digitalWrite(13, LOW);
    else digitalWrite(13, HIGH); 
  }
}

```
