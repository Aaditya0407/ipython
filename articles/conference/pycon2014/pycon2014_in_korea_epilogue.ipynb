## 행사 피드백

### 진행이 잘된 점 

1. 행사장 곳곳에 자원봉사자 및 운영위원회 분들이 계셔서 어렵지 않게 다닐 수 있었음
2. 홈페이지가 참 **Interactive** 했음(NDC13 덕택!)
3. **재미있는 주제**가 참 많아서 **선택의 즐거움** 및 고통
4. 홈페이지 접속해서 **인증**하고 **발급 코드 입력**하는 부분이 인상적
5. 식사도 괜찮았음

### 매끄럽게 진행되지 못한 아쉬운 점 몇 가지

1. 8월 1일에 참가 등록 페이지가 올라오지 않았음(이때 박현우님의 commit은!!)
2. 프로그램(Start, Running)의 구분이 크게 의미가 없었던듯
3. 나눠준 용지의 시간과 맞지 않음(**홈페이지 실시간 업데이트로 대체**)
3. 행사장이 연속으로 붙어있는 것이 아니어서 약간 당황했음
4. **식사**하러 가는데 산넘어 물건너 도착했더니 등에 **땀**이 흥건했지만 **식사**는 **만족**!
