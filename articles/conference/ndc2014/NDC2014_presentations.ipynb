### 훌륭한 유저 동향 분석 시스템 시리즈

- 던파팀에서 1가지 주제를 가지고 왠 세션을 3개나 맡았나 했더니 그럴만 했다. 정말 아이디어가 훌륭하다.
- 처음에는 유저와 반응하는게 부담이었을텐데 잘 이겨냈고 시스템도 무리하게 진행하지 않고 단계적으로 진행이 됐다.
- 관찰 -> 문제점 파악 -> 문제 수정 -> 피드백, 이 선순환 구조 반복
- 처음은 문제를 관찰한다. 그 문제를 이야기 한다. 그 사람은 그 문제를 해결하기 위해서 이야기를 했던거다. 그 문제를 해결하려고 하는 의지조차 없었다면 이야기를 꺼내지 않았을 것이고 침묵했을 것이다. 그 문제를 이야기 하는게 정말 어렵다는걸 이해한다. 
- 어떻게 상대방이 불편하지 않게 전달할 수 있을까? 이게 인간 관계의 본질이 아닐런지.. 그 사람은 이러이러한걸 고치면 더 좋은 관계가 될 수도 있다고 이야기 한 거지만 상대방은 그렇게 받아들이지 않을 수도 있다. 말하기 전에 한 번 더 생각을 해보자. 
- 문제점이 있다고 인식당한 사람의 대응 방식도 관건이다. 긍정적으로 받아들여 좀 더 발전적인 방향으로 가느냐? 아니면 부정적으로 받아들여 소원한 방향으로 가느냐? 인생을 살면서 참 어려운 주제같다.
