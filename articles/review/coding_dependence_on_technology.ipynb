### 1.2 역사를 통한 배움

- 언어 설계자의 의도를 이해하자
- 어떤 언어를 배워야 하는지는 아무도 모른다.
- 언어에 의존하지 않는 보편적인 지식의 습득

#### 정리

- 특정 언어에 국한된 지식이 아니라, 보다 **보편적인 지식을 습득**할 수 있도록 하고 있음
- 이를 위해 **비교를 통한 배움**과 **역사를 통한 배움**이라는 2가지 방법을 사용
- 비교를 통한 배움
  - 특정 언어로 프로그래밍을 배우는 것이 아니라, 다수의 언어를 비교해 가면서 학습하는 것을 의미
  - 이를 통해, 무엇이 언어에 따라 다르고 무엇이 공통적인지 배울 수 있음
- 역사를 통한 배움
  - 언어가 어떻게 바뀌었고 바뀌기 전에는 어떤 의문점이 존재했는지 학습하는 것을 의미
  - 이를 통해, 언어가 가지고 있는 다양한 기능이 **'왜'** 탄생했는지 배울 수 있음

#### 정리

- 언어는 도구다. 어떤 언어가 자신의 목적에 적합한지는 **그 언어를 사용할 때 자신이 어느 정도의 생산성을 발휘할 수 있는지로 판단**할 수 있음

- 다른 사람이 말하는 것에 혹하지 말고, 좋은 도구를 현명하게 선택하여 적재적소에 사용하도록 하자

#### 칼럼: 이해력을 확인하기 위해서는 결과물(Output)을 확인한다

>- 여러분이 어떤 것을 공부해서 '이해했다'고 치자. 그것은 '정말 이해했다' 일까 아니면 '이해했다라고 착각했다'일까?
- **이해한 것이 맞는지를 확인**하기 위해서는 **Output이 필요**하다. 자신이 이해한 것을 정리해서 그것을 제3자에게 검증을 받는 수 밖에 없다. 영어를 공부했다면 그 영어를 사용해 다른 사람의 반응을 확인하지 않는 한, 자신이 영어를 정말 이해했는지 알 수 없음
- 프로그래머는 이런 점에서는 **혜택**
- 예르 ㄹ들어, 문서라는 결과물은 틀린 곳이 있어도 누군가 시간을 내어 검토해주지 않으면 잘못을 모르는 경우가 발생함
- 그러나 프로그래밍 언어의 결과물은 언어 처리계가 **틀린 곳을 자세히 지적**해줌
- 바쁜 사람과는 달리, 언제든 편한 시간에 몇 시간이라도 시간을 할애해줌
- 에러 메시지가 발생하면 '왠 에러가 떴어!'라고 당황하거나 무서워서 그냥 무시하는 사람도 많음. 
- 이는여러분이 만든 것을 언어 처리계가 열심히 읽어서 '여기를 모르겠어'라고 대답해주고 있는 것
- 그것을 이해하려 하지 않고서는 언어 처리계와 제대로 된 커뮤니케이션을 할 수가 없음

### 3.4 중위 표기법

- 구문 해석기
#### 칼럼: 무엇을 배우면 좋을지 모르는 이유


>- '무엇을 배우면 좋습니까?' 이는 자주 듣는 질문
- 이 질문에 대답하기 전에 우선 한 가지를 물어보고 싶다. '여러분은 무엇을 만들고 싶은가요?' **목적이 명확하지 않으면 최선의 방법 또한 가르쳐줄 수 없다.**
- 정보는 방대하다. '어쨋든 전부 다 한 번은 공부해보자'는 전략은 어리석다. '무엇을 배울지, 무엇을 배우면 좋을지'를 정할 때 필연적으로 지표가 될 수 있는 것이 '무엇을 만들고 싶은가?', 즉 목적을 명확히 하는 것
- 그리고 목적을 달성하기 위해 필요한 것부터 배우면 됨
- '무엇을 만들까?'를 결정할 수 없다면 혹시 처음부터 완벽한 것을 만들려고 생각하고 있는 것은 아닌지 고민해 보자.
- 누구도 생각한 적이 없는, 모두를 만족시키는 것을 만들려고 생각하고 있지는 않은가?
- 처음부터 굉장한 것을 만들려고 생각해서 오히려 손도 대지 못하고 있다면, 시간이 흘러도 굉장한 것을 만들 수 없다.
- 우선은 간단한 것도 괜찮으니, 무엇이라도 만들어보자. 그 과정에서 '나는 이것을 할 수 있어', '이것은 아직 어려우니까 공부할 필요가 없어' 등을 알 수 있게 됨
- 이 과정을 반복해가면 보다 복잡한 것을 만드는 능력이 몸에 베개 됨


- 규칙간 마찰

#### 정리

- '1 더하기 2에 3을 곱한다'는 똑같은 처리도, 언어어 따라 표현 방법이 틀림
- 그러나 구문 트리로 표현하면 거의 동일함
- 이들 언어에 차이가 생긴 것은 '어떤 문자열을 쓰면 어떤 구문 트리가 생기는가'라는 규칙이다. 이것이 문법이다.
- FORTH나 LISP은 규칙이 적은 것을 중시함
- 그러나 시장이 요구한 것은 규칙이 적거나 간단한 것이 아니었음
- FORTRAN은 '\*가 +보다 우선 순위가 높다' 등의 정해진 규칙을 대량으로 도입해서 '다가가기 쉬운 작성법'을 중시함
- 그런데 그런 설계 방침의 성공으로, FORTRAN 방식이 LISP나 FORTH보다 많은 사람에게 사랑받게 됨
- 프로그램을 짤 때 '왜 이렇게 다가가기 어려운 작성법이 필요한지' 생각한 적이 있을것임
- 현재 대부분의 언어는 FORTRAN 식의 '다가가기 쉬운 작성법'을 목표로 하고 있음
- 하지만 모순없이 해석할 수 있는 문법을 만들어내는 것은 어려운 작업
- 특히 나중에 새로운 문법을 추가할 때 기존 문법과 마찰되지 않도록 하는 것은 더욱 그렇다.
- 이 때문에 현실의 프로그래밍 언어에는 이해하기 이러여누 작성법이 여전히 존재하는 것

#### 칼럼: 구체적인 지식과 추상적인 지식

>- '언어 X로 Y를 하려면 어떻게 해야 하나?' 이런 구체적인 how-to 지식은 당신의 생산성을 바로 높을 수 있음

- 즉, **응용 범위가 제한된 구체적인 하우투는 점점 그 가치를 잃어감**
- 하우투뿐만 아니라 **의식적으로 응용범위가 넓은 추상적 개념을 배울 필요**
- 반면, 추상적인 지식을 배워도 여러분이 가지고 있는 경험과 연관시지키지 못하면 응용할 수 없음
- 벚꽃이 가지고 싶다고 꽃이 피어있는 가지를 잘라와도 꽃이 죽어버리면 그만
- 매년 꽃을 피게 하기 위해선 **뿌리가 필요**
- **지식에 뿌리가 있는지 없는지**는 **구체적으로 예를 든 설명이나 구체적으로 구현이 가능한지로 확인 가능**
- 뿌리가 없는 지식은 응용에 도달하지 못하고, 들어온 지식을 다시 토해낼 뿐
- **상황에 맞게 지식을 활용할 수가 없는 것**

#### 칼럼: 이빨로 씹다

>- 큰 고기를 한 입에 먹을 수는 없음
- 우선은 입에 넣을 수 있는 크기로 잘라서, 이빨로 씹은 후 삼켜야 함
- '이해'란 것도 같다.
- 어려운 개념이나 복잡한 시스템, 읽숙하지 않은 분야를 한 번에 모두 이해하는 것은 어렵다.
- 우선은 **작은 정보 단위로 잘라서, 하나씩 씹어 자신의 것**이 되도록 흡입해야 함
- 그러나 방대한 정보 중에서 무엇을 선택하면 좋을까? 어떤 정보가 중요하고 어떤 정보가 중요하지 않을까? 어떤 정보가 중요한지 판단하기 위해서는 **대상을 깊이 이해**할 필요가 있음
- 지인 중에 전문가가 있다면 그 사람에게 배우는 것도 한 가지 방법
- 하지만 주위에 그런 사람이 없다면 인터넷을 검색해서 설명을 발견했다고 해도 해당 저자가 정말 전문가인지, 아니면 전문가인 척하는 것인지 어떻게 알 수 있을까?
- 예를 들어, 저자가 직접 쓴 자료나 소스가 가장 상세하다고 치자. 그러나 '어렵다', '양이 많다', '영어라서 읽을 수 없다' 등의 변명으로 다른 사람이 쓴 쉬운 설명을 찾으려고 함
- 그것은 **'원래 고기는 너무 크고 먹기가 어려워'** 라고 하며, **안전한지도 확인도 안하고 다른 사람이 만든 다진 고기를 먹는 것**과 동일
- 이해는 간다. 저자도 방대한 정보를 앞에 두고 좌절한 것이 한 두번이 아니다. 그럴떄 사용할 수 있는 3가지 전략이 있다.


#### 방대한 정보 빨리 흡수하기 3가지 전략

1. 필요한 부분부터 흡수한다.
2. 대략적인 부분을 잡아서 조금씩 상세화함
3. 끝에서부터 차례대로 베껴간다.

#### 정리

- 프로그램도 실패하는 때가 있다는 것
- 그리고 그 실패를 어떻게 전달할지에 대해 배움
- 실패를 전달하는 방법 크게 2가지
  - 반환값으로 알린다
  - 실패하면 점프한다.
- 반환값으로 알린다 전략에는 반환값을 잊어버려서 실패를 놓칠 수 있는 문제 발생
- 그래서 모든 실패를 잡을 수 있는 '실패하면 점프한다'는 전략이 오랜 시간 연구되어 왔음
- 이것이 지금의 Java, C++, Python, Ruby 등 많은 언어가 지원하고 있는 '예외 처리'의 구조
- 한편, 예외 처리에도 몇 가지 문제가 존재
- 하나는 함수의 출구가 여러 개로써 '짝을 이루는 처리'을 바르게 짝을 처리하는 거싱 어렵다는 문제
- 다른 하나는 함수가 어떤 에외를 던질지 그 함수의 코들르 확인해도 알 수 없다는 문제
- 2010년 현재, Google 코딩 규약에서는 C++로 예외를 사용하는 것을 금지하고 있음
>- 에외를 사용하는 이점은 비용 대비 매우 효율적이라는 것. 신규 프로젝트에서는 더욱 그렇다
- 그러나 기존 코드에 예외를 도입하려고 하면, 의존 관계가 있는 모든 코드에 영향을 끼칠 수 있음
- (중략)Google의 기존 C++ 코드 대부분은 예외 처리를 할 수 있도록 만들어지지 않음
- 따라서 예외를 발생시키도록 하는 신규 코드를 추가하는 것은 어렵다


#### 칼럼: 필요한 부분부터 흡수한다

>- 방대한 정보 앞에서 좌절하고 있을 때 어떻게 하면 좋을까?
- 그 첫번째 방법이 '필요한 부분부터 흡수한다'이다.
- 책이나 자료 전체가 동일한 정도로 중요하다고 말할 수 없음
- 목적이 명확하고 목적 달성을 위해서 어디를 읽어야 할지 알고 있다면 다른 페이지는 신경 쓰지 말고 바로 그곳을 읽도록 함
- 전체 모두 읽지 않는 것이 께름칙한가? 하지만 좌절하고 전혀 읽지 않는 것보다는 낫다. '전부 읽지 않으면' 이라는 완벽주의가 배우고자 하는 동기를 짓누르고 있다면, 버려버리는 것이 낫다. **동기는 매우 중요**하다.
- 이 전략을 사용하기 위해선느 읽고 싶은 부분이 어디인지 **대략적으로 전체적인 구조를 파악**하고 있어야 함
- 만약 그게 어려우면 다음 전략인 **'대략적인 부분을 잡아서 조금씩 상세화'** 한다를 시험해보도록 하자

#### 칼럼: 다른 언어의 스코프

>- 이 장의 샘플 코든느 대부분 Perl을 사용함
- Perl에서 아무것도 선언하지 않고 변수를 선언하면 전역 스코프
- local을 붙이면 동적 스코프(1991년 Perl 4부터)
- my를 붙이면 정적 스코프(1994년 Perl 5부터) 등 스코프 설명에 편리한 특징을 가지고 있음
- 현재는 규칙에 맞지 않는 코드를 금지하는 모드(use strict)로 실행하면 local로 선언한 동적 스코프 변수나 패키지명을 지정하지 않은 전역 변수는 컴파일 시에 에러를 일으킴
- 다른 언어는 어떨까/ 역사적으로는 1958년에 만들어진 초기 LISP는 동적 스코프였음
- 그리고 1975년에 만들어진 LISP의 일종인 Scheme은 정적 스코프를 채용함
- 1994년에 만들어진 Javascript에서는 Perl과 동일하게 아무것도 선언하지 않으면 전역 스코프, var를 붙이면 정적 스코프
- 1991년의 Python과 1995년의 Ruby에서는 아무것도 선언하지 않아도 정적 스코프
- 전역 대응표는 모두의 공용 공간이기 때문에 이를 어지럽히지 말고 변수의 영향 범위를 작게 하여 이해하기 쉬운 코드를 만들 수 있음

- 내포 함수의 문제점
- 외부 스코프에 재귀속되는 문제

#### 정리

- 이름의 유효 범위를 제한하는 것이 얼마나 중요한지 배움
- 이름의 유효 범위는 2013년 현재 **정적 스코프**가 **주로** 사용되고 있음
- 변수는 거의 모든 언어에서 사용하고 있어서 '어떤 언어에서도 같다', '옛날부터 바뀌지 않았다'라고 착각하기 쉬움
- 하지만 언어에 따라 다양한 차이가 있어서 보다 개선된 방법을 위해서 아직도 다양한 논의가 이루어지고 있음
- 이 장에서는 하나의 예로 Ruby 1.9나 Python 3.0의 변경사항에 대해 설명함
- 한편 동적 스코프는 많이 사용되고 있지는 않지만, 아직까지 무용지물이 된 건 아님
- 정적 스코프에서는 스코프가 소스코드 상에 한 묶의 영역으로 존재하지만, 동적 스코프에서는 스코프에 들어가서 나갈 때까지 시간축 상에 존재하는 범위
- 이와 같은 특징을 가지고 있는 것이 몇 가지 있음
- 예를 들어 '어떤 처리를 하고 있는 중간에 일시적으로 변수값을 바꾸고 나중에 되돌린다'는 처리를 하고 있다면 사전에 동적 스코프를 만들고 있는 것과 같음
- 또한 예외 처리도 동적 스코프와 닮음
- 어떤 함수내에서 예외를 던질 때의 처리가 호출처에 있는 함수의 try/catch로 인해 영향을 받음
- 객체 지향에서 private 선언 등의 접근 지시자도 접근 가능한 범위를 제한한다는 점에서는 스코프와 닮음
- 접근 가능한 범위가 클래스 내로 제한되는 private에 비해 protected는 자식 클래스까지 범위가 넓어짐
- 이것은 호출처에서 변경한 것이 호출한 곳에 영향을 끼치는 동적 스코프와 닮음
- 어느 쪽도 '영향 범위가 코드 상의 한 곳에 모여지지 않는다'는 단점 존재
- 예를 들어 Java는 정적 스코프 언어이지만 클래스는 소스코드 어디서든 참조할 수 있음
- 즉, 클래스는 전역 스코프라고 할 수 있음
- 클래스는 이름이 계층적으로 이루어져 있고, 임포트하지 않으면 사용할 수 없어서 전역 변수가 가지고 있던 '충돌' 문제를 피하고 있음
- 그러나 전역 변수도 클래스의 static 멤버도 '소스코드 어디서든지 바꿀 수 있다'는 점에서 같다. 편리하게 사용할 수 있지만, 남용하면 이해하기 어려운 코드가 되버림
- 스코프는 이해하기 쉬운 코드를 만들기 위한 핵심 개념으로 다양한 부분에서 계속적으로 사용되고 있음

## 9. 컨테이너와 문자열

### 9.1 다양한 종류의 컨테이너

### 9.2 왜 다양한 컨테이너가 존재할까?

- 배열과 연결 리스트
- 연결 리스트의 장단점
#### 칼럼: O 기법? 계산 시간과 데이터량의 관계를 간단히 나타내는 것

>- 계산에 걸리는 시간과 데이터량의 관계에 대한 간단한 기법 소개
- 배열에 삽입하는 경우와 같이 '데이터량 n이 2배가 되면 계산 시간도 2배가 됨'는 성질을 O(n)이라 쓰고 'n 오더(order)' 라고 부름
- 반면, 연결 리스트에 삽입하는 경우와 같이 '데이터량 n이 2배가 되도 게산 시간은 바뀌지 않는다'는 성질을 O(1)이라 쓰고 '정수 오더'라고 부름
- 이외에도 데이터량이 2배, 3배가 됐을 때 계산 시간이 4배, 9배로 증가하는 O(n2)이나 데이터량이 2배가 됐을 떄 증가하는 계산 시간과 데이터량이 2배가에서 4배로 됐을 때 증가하는 계산 시간이 같아지는 O(log n) 등이 있음
- 큰 데이터를 가지고 for문으로 처리할 때 O(n)이 되고, 이중 루프를 쓰면 O(n2)가 됨
- n이 커지면 다음 순으로 계산 속도가 빠름
- O(1) < O(log n) < O(n) < O(n2)
- 이 표기 방법은 함수의 증가 방식을 매우 간단히 표현하고 있어서 n2 + 2n + 1이든 3n2이든 O(n2)가 됨. 즉 상세한 부분은 생략하고 표현하고 있음
- '오더 엔' 또는 '오 엔'이라고 부르는 경우도 있음

#### 칼럼: 끝에서부터 차례대로 베껴간다

- 방대한 정보 앞에 좌절될 때 어떻게 하면 좋을까? 제 6장의 칼럼 '이빨로 씹다'에서 3가지 방법을 소개함
- 첫번째는 '필요한 부분부터 흡수한다'
- 두번째는 '대략적인 부분을 잡아서 조금씩 상세화한다'
- 둘 다 성공하지 못한다면 최후의 수단으로 '끝에서부터 차례대로 베껴간다'가 있음
- 명확히 '하고 싶은 것', '조사하고 싶은 것'이 없이 '대충 읽으면' 읽은 내용이 뇌를 그냥 스쳐 지나갈 뿐
- 이런 상태에서 어떻게 배울까를 고민한다고 해도, 판단을 위한 지식 자체가 없기 때문에 무의미
- 그래서 지식의 밑바탕을 만들기 위해서 교과서를 그대로 베껴 씀
- 이것이 '베끼기'라 불리는 기술. 지식이 없는 상태에서 고민하는 것은 무익하기 떄문에 우선 아무것도 생각하지 않고 지식을 복사하는 것
- 이 이상의 방법은 없다. 저자는 시간을 정해서 '25분간 어디까지 베낄 수 있는지' 도전하는 것을 좋아함
- 분량으로 나누는 것도 좋은 방법
- 중요한 것은 **간격을 적절히 해서 목표를 이루었다는 만족감**을 얻을 수 있도록 하는 것
