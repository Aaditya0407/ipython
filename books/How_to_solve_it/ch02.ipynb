# 제 2부 어떻게 문제를 풀 것인가?

- 2장이 이 책의 가장 핵심
- 나머지 장도 시간되면 정리하겠음
- 현재는 1,2장 읽어보고 나머지 장은 훑어만 봄

## 책을 어떻게 하면 효율적으로 읽을 수 있을까?

- 많이 생각해 봤는데 포토 리딩 방법이 좋음


1. 책의 목차 보기
2. 개요사, 후기 등은 꼭 읽어보기(여기에 저자의 노력이나 가치관 등이 있기 때문에 전체 책의 뉘앙스를 파악하기 아주 좋음)
3. 한장씩 넘기면서 큰 주제들만 보면서 빠르게 훑기
4. 이번에는 소주제들도 보면서 빠르게 훑기
5. 머리속에 어느 정도 목차들이 들어왔다면 자세히 읽으면서 중요한 곳에 밑줄치기
6. 읽으면서 내 생각을 책 옆에 정리하기(나와 의견이 다른 부분, 일치하는 부분)
7. 책을 다시 한 번 빠르게 훑어보면서 정리하기(내가 글씨 쓴 부분, 밑줄 친 부분 위주)
8. 책 내용을 정리해서 공개하기(공개하려면 아무래도 한 번이라도 더 보게 되고 오타있는지 확인하게 되고 공개하고도 또 틀린 부분은 없는지 읽어보게 되서 저절로 반복학습이 됨)
9. 필요할 떄 내가 쓴 글을 검색해서 활용하기

## 2.1 눈에 익히기

- "어디서 시작해야 할까요?" 
  - 문제에 대한 진술을 바탕으로 시작
- "어떻게 할까요?"
  - 문제를 가능한 한 분명하고 생생하게 전체적으로 머릿속에 그려 보아라
  - 잠시 세세한 것에는 관심X
- 그렇게 함으로써 어떤 이득이 있습니까?
  - 여러분은 문제를 이해하여야 하고 문제에 친숙해져야 하며, 마음 속에 그 목적을 새겨 두어야 함
  - 문제에 관심을 쏟게 되면 기억을 자극할 수도 있고 관련된 요점을 생각해 내는 데 도움이 될 수도 있음

## 2.2 보다 이해를 잘하기 위한 활동

- "어디서 시작해야 할까요?"
  - 다시 문제에 대한 진술을 바탕으로 하여 시작
  - 문제의 진술이 분명해지고 여러분들의 마음 속에 잘 새겨져서 전혀 잊어버릴 걱정이 없을 때 시작
- "어떻게 할까요?"
  - 문제의 핵심을 뽑아내어라.
  - "증명 문제"에서는 **가정**과 **결론**이 핵심
  - "답을 구하는 문제"에서는 **미지인 것**, **자료**, **조건**이 핵심
  - 문제의 핵심 부분을 샅샅이 조사하여 하나 하나씩 차례차례 생각해 보고, 
  - 여러 가지로 조합하여 생각해 보기도 하고, 
  - 각 부분을 서로 관련시켜 보기도 하며, 
  - 개개의 것을 문제 전체와 관련시켜 보기도 하라
- "그렇게 함으로써 어떤 이득이 있습니까?"
  - 문제를 푸는데 중요한 역할을 할 것 같은 **세목**을 **미리 생각**해 두고 **명확**하게 해 놓아야 함

## 2.3 도움이 되는 아이디어 찾기

- "어디서 시작해야 할까요?"
  - 핵심 부분들이 뚜렷하게 정리되고 분명하게 인식되었을 때에, 내용에 대한 기억이 가능할 듯할 때 시작
- "어떻게 할까요?"
  - 문제를 여러 측면에서 생각해 보고 이전에 얻은 지식과 연결이 되는가 살펴보라([메타 인지](https://gist.github.com/re4lfl0w/9c5cd30f6f0b0f097243)와 관련됨)
  - 각 부분의 차이를 부각시켜 보라
  - 여러 부분을 검토하여 보라
  - 같은 부분을 거듭 여러 가지 방법으로 조사하여라
  - 세세한 부분을 여러 가지로 조합해 보고 여러 측면에서 살펴보라
  - 개개의 세세한 부분에서 새로운 의미를 찾아내고 전체를 새롭게 해석해 보라
  - 이전에 습득한 지식과 연결되는지 살펴보라
  - 과거에 유사한 상황에서 여러분에게 도움이 되었던 것이 무엇이었는가를 생각해 보라
  - 여러분들이 조사한 것 가운데에서 친근한 것이 있는가 생각해 보라
- "무엇을 파악할 수 있을까요?"
  - 한 눈에 여러분에게 바로 목표까지 이르는 길을 보여 줄, 도움이 될 만한 아이디어, 아마도 결정적인 아이디어를 파악해야 함
- "아이디어가 어떻게 도움이 될까요?"
  - 그러한 아이디어는 해답에 이르는 길 전체 또는 일부를 보여줄 것
  - 아이디어는 여러분들에게 어떻게 나아갈 수 있는가를 다소 분명하게 제시해 줌
  - 아이디어란 대체로 완전한 경우가 보통
- "불완전한 아이디어라면 어떻게 할까요?"
  - 그러한 것들도 고려해 보아야 함
  - 만약 도움이 될 만해 보이면 좀 더 오랫동안 그것을 생각해 보아야 함
  - 그것이 믿을만하다는 생각이 들면 여러분을 어디까지 인도해 갈 수 있는가를 확인해 보아야 하고 그 상황을 재고해 보아야 함
  - 도움이 되는 아이디어 덕택에 상황이 바뀜
  - 여러 측면으로부터 그러한 새로운 상황을 숙고해 보고 이전에 습득한 지식과 연결이 되는가를 살펴보라
- "다시 그렇게 함으로써 어떤 이득이 있습니까?"
  - 운이 좋아서 다른 아이디어를 생각해 낼 수도 있음
  - 아마도 그 새로운 아이디어로 인해 여러분들은 즉시 해답을 찾아낼 수도 있을 것
  - 그 아이디어 외에 1~2가지 보다 도움이 될 만한 아이디어가 필요 할지도 모름
  - 아이디어 때문에 방황할 수도 있음
    - 모든 새로운 아이디어
    - 그보다 못한 아이디어
    - 모호한 아이디어를 다소 정확하게 해주는 또는 보다 미진한 아이디어를 수정해 주는 보충적인 여러가지 아이디어에 고마운 마음을 가져야 함
    - 비록 잠시 동안 어떤 가치있는 새로운 아이디어도 가지고 있지 못하다고 하더라도 문제에 대한 여러분의 인식이 보다 완벽해지고 보다 조리가 서며, 또한 보다 일관성이 있고 보다 균형이 잡히게 되면 여러분들은 고마워해야 할 것

## 2.4 계획의 실행

- "어디서 시작해야 할까요?"
  - 여러분들을 해답으로 인도하는 행운의 아이디어로부터 시작하라
  - 주요한 관계를 확실히 이해했다고 확신할 때, 그리고 빠질 수도 있는 그리 중요하지 않은 부분들을 보충할 자신이 생겼을 때 시작하라
- "어떻게 할까요?"
  - 파악한 것을 아주 확실히 하라
  - 여러분들이 이전에 할 수 있다고 여겨 온 모든 대수학적 기하학적 조작을 가능한 자세히 수행해 나아가라
  - 형식적 추론 또는 직관적 통찰 또는 가능하다면 그 2가지 방법으로 매단계의 정확성을 확인하라
  - 만약 문제가 복잡하다면 "큰" 단계, "작은" 단계로 구분해서, 
  - 큰 단계가 여러 개의 작은 단계로 이루어지도록 할 수도 있음
  - 먼저 큰 단계를 점검하고 다음에 점차 작은 단계를 점검해 보도록 하여라
- "그렇게 함으로써 어떤 이득이 있습니까?"
  - 매단계가 의심할 여지없이 정확한 그러한 해답을 제시할 수 있음

## 2.5 반성

- "어디서 시작해야 할까요?"
  - 세세한 부분에 대한 완전하고 정확한 해답으로부터 시작하여라
- "어떻게 할까요?"
  - 여러 측면에서 해답을 숙고해 보고 이전에 습득한 지식과 연결이 되는가 살펴보아라(TestCase 작성하라는 이야기)
- 해답의 세세한 부분을 숙고해 보고 할 수 있는 한 단순화해 보도록 하여라
- 전체적인 해답을 일별해 보아라
- 해답의 크고 작은 부분을 적절히 수정해 보도록 하여라
- 전체 해답을 개선해 보려고 하고, 해답을 직관적으로 생각해 보고 가능한 한 자연스럽게 여러분들이 이전에 습득한 지식에 맞추어 보도록 하여라
- 해답에 이르게 된 방법을 자세히 음미해 보고, 그 요점을 살펴보도록 하고, 다른 문제에 그것을 이용해 보도록 하여라
- 결과를 자세히 음미하고 그것을 다른 문제에 사용하도록 해 보아라
- "그렇게 함으로써 어떤 이득이 있습니까?"
  - 새롭고 보다 좋은 해답을 얻을 수도 있으며, 새롭고 흥미있는 사실을 발견할 수도 있을 것
  - 여하튼 이러한 방법으로 **풀이를 개관**하고 **음미하는 습관**을 들이게 되면, 여러분들은 **잘 정돈**되어 **곧바로 사용될 수 있는 지식**을 얻게 될 것이며, **문제를 해결하는 능력**이 **개발**되게 될 것
